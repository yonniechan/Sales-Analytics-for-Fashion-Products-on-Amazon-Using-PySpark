# Clean products.json¶

```python
target_keys_product = ['title', 'rank', 'brand', 'asin']
```

```python
fashion_product = []
with open('./Data/meta_AMAZON_FASHION.json') as f:
    for line in f:
        obj = json.loads(line)
        valid = True
        for target_key in target_keys_product:
            if target_key not in obj:
                valid = False
                break
        
        if valid and len(obj['title']) >= 300:
            valid = False
        
        if valid:
            n_obj = {target_key: obj[target_key] for target_key in target_keys_product}
            n_obj['title'] = n_obj['title'].replace("\"","")
            
            n_obj['rank'] = "".join(n_obj['rank'])
            
            n_obj['rank'] = n_obj['rank'][:n_obj['rank'].find(" (See Top")]
            n_obj['rank'] = n_obj['rank'].replace(" ", "")

            rank = n_obj['rank'].split('in')
            rank[1] = "in".join(rank[1:]).replace("amp;","")
            if rank[1] not in ["Clothing,ShoesJewelry", "Clothing,Shoes&Jewelry"]:
                continue
            if ">#" in rank[0]:
                rank[0] = rank[0].replace(">#", "")
            n_obj['rank'] = int(''.join(rank[0].split(',')))
            
            fashion_product.append(n_obj)
```

```python
len(fashion_product)
```

```python
product_df = pd.DataFrame.from_records(fashion_product)
```

```python
product_df.info()
```

```python
product_df.to_csv('./Data/product_fashion_cleaned.csv', index = False)
```

# Clean review.json

```python
target_keys_review = ['overall', 'asin', 'reviewText']
```

```python
fashion_review = []
with open('./Data/AMAZON_FASHION.json') as f:
    for line in f:
        obj = json.loads(line)
        valid = True
        for target_key in target_keys_review:
            if target_key not in obj:
                valid = False
                break
        if valid and "verified" in obj and obj['verified']:
            n_obj = {target_key: obj[target_key] for target_key in target_keys_review}
            n_obj['vote'] = 0 if 'vote' not in obj else int(obj['vote'])
            
            n_obj['reviewText'] = n_obj['reviewText'].replace("\"", "")
            
            
            fashion_review.append(n_obj)
        
```
